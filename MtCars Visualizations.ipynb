```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the mtcars dataset
mtcars = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')
```

```python
# 1. Line Chart: Relationship between MPG and Horsepower

# Create the line chart
plt.plot(mtcars['mpg'], mtcars['hp'])

# Add labels and title
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Horsepower (hp)')
plt.title('Relationship between MPG and Horsepower')

# Show the plot
plt.show()
```

```python
# 2. Scatter Plot: Relationship between Weight and MPG

# Create the scatter plot
plt.scatter(mtcars['wt'], mtcars['mpg'])

# Add labels and title
plt.xlabel('Weight (wt)')
plt.ylabel('Miles per Gallon (mpg)')
plt.title('Relationship between Weight and MPG')

# Show the plot
plt.show()
```

```python
# 3. Horizontal Stacked Bar Chart: Total number of cars by cylinder and transmission type

# Group the data by cylinder and transmission type
car_counts = mtcars.groupby(['cyl', 'am'])['model'].count().unstack()

# Create the horizontal stacked bar chart
car_counts.plot(kind='barh', stacked=True)

# Add labels and title
plt.xlabel('Number of Cars')
plt.ylabel('Cylinders')
plt.title('Total Number of Cars by Cylinder and Transmission Type')

# Show the plot
plt.show()
```

```python
# 4. Vertical Stacked Bar Chart: Total number of cars by transmission and gear type

# Group the data by transmission and gear type
car_counts = mtcars.groupby(['am', 'gear'])['model'].count().unstack()

# Create the vertical stacked bar chart
car_counts.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Transmission Type')
plt.ylabel('Number of Cars')
plt.title('Total Number of Cars by Transmission and Gear Type')

# Show the plot
plt.show()
```
